```python
#  Prepare python programm for data cleaning process to removing unnecessary data
```

```python
# loading libraries
import pandas as pd
```

```python
# loading data
df = pd.read_excel("Yoshops_Order.xlsx", header=1)

df.head(5)
```

```python
# checking null values
df.isna().sum()
```

```python
df.columns
```

```python
# rows which cotians null values
df[df[['Address 1', 'Adress 2','pincode']].isnull().any(axis=1)]
```

As you see these columnns are full of misinformation so we are dropping it 

```python
# dropping null values
df.dropna(inplace=True)
```

```python
# checking duplicates
df[df.duplicated()]
```

```python
#dropping duplicates

df.drop_duplicates(inplace=True)
```

```python
# saving the cleaned data to excel files 
df.to_excel("cleaned_data.xlsx", header=True, index=False)
```
