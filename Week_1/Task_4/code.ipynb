# Python program for Prepare Yoshops Survey and Order excel charts Like = Pie Chart and Bar Chart Weekly, Monthly and Yearly Reports.


```python
#loading libraries
import pandas as pd
import matplotlib.pyplot as plt

survey_df=pd.read_excel("Yoshops Survey.xlsx")
```

```python
survey_df.head()
```

```python
# Convert 'Submitted Time' to datetime
survey_df['Submitted Time'] = pd.to_datetime(survey_df['Submitted Time'], format='%d-%m-%Y')
```

```python

# Set 'Submitted Time' as the DataFrame index
survey_df.set_index('Submitted Time', inplace=True)
```

```python
# Prepare monthly report - Pie Chart
monthly_counts = survey_df.resample('M').size()
plt.figure(figsize=(12, 6),dpi=200)
monthly_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Monthly Survey Responses')
plt.ylabel('')
plt.show()
```

```python
# Prepare yearly report - Bar Chart
yearly_counts = survey_df.resample('Y').size()
plt.figure(figsize=(8, 6),dpi=200)
yearly_counts.plot(kind='bar')
plt.title('Yearly Survey Responses')
plt.xlabel('Year')
plt.ylabel('Number of Responses')
plt.show()
```

```python
# Prepare weekly report - Bar Chart
weekly_counts = survey_df.resample('W').size()
plt.figure(figsize=(10, 6),dpi=200)
weekly_counts.plot(kind='bar')
plt.title('Weekly Survey Responses')
plt.xlabel('Week')
plt.ylabel('Number of Responses')
plt.show()
```
